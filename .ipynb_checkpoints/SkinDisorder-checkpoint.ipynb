### Domain Analysis :

- **Erythemato-squamous diseases (ESDs)** are a group of common skin conditions that cause inflammation and scaling, usually on the face and scalp.
- The six important types of ESD are including **psoriasis, seborrheic dermatitis, lichen planus, pityriasis rosea, chronic dermatitis and pityriasis rubra pilaris**. 
- For instance, the **psoriasis** is the most common disease of ESD which approximately, **2 to 3% of population are affected**.
- The skin appears red due to the shedding of skin cells and other symptoms include itching, scaling, and involvement of the oral mucosa.
- ESDs are often caused by genetic or environmental factors.
- All of them have similar clinical and histopathology features such as erythema and scaling which makes it difficult to detect them even with biopsy, biopsy is necessary to achieve a correct diagnosis. 
- Usually, clinicians detects these diseases using some clinical features, such as the degree of erythema and scaling, border of lesion, the presence of itching, koebner phenomenon, the formation of papule, involvement of oral mucosa, elbows, knees and the scalp and family history
- The six phenotypes of Erythmato-squamous Diseases are the major challenge in dermatology.
- These diseases are frequently seen in the outpatient department of dermatology. 
- All of ESD disease are same in terms of erythema and scaling, but when examine carefully some patient have the typical clinical feature of disease.
- Due to overlapping between sign and symptom of ESD, differential diagnosis is very difficult. 
- Automatic detection of these group of disease could help physician for decision making.

### Meaning of the every feature available in the dataset

##### **Clinical Features:**

- **Erythema:** Redness of the skin due to increased blood flow, often seen in inflammatory conditions.
- **Scaling:** The shedding of dead skin cells, creating flakes on the surface, common in psoriasis, seborrheic dermatitis, etc
- **Definite borders:** Clear, well-defined boundaries of lesions, often seen in conditions like psoriasis or ringworm.
- **Itching:** A common symptom of many skin conditions, including eczema and psoriasis.
- **Koebner Phenomenon:** The appearance of new skin lesions at sites of trauma or injury in patients with certain skin disorders like psoriasis or lichen planus.
- **Polygonal Papules:** Small, raised, flat-topped lesions with a polygonal shape, seen in conditions like lichen planus.
- **Follicular Papules:** Raised bumps that appear around hair follicles, commonly seen in acne or keratosis pilaris.
- **Oral Mucosal Involvement:** Lesions that appear on the mucosal surfaces inside the mouth, such as in lichen planus.
- **Knee and Elbow Involvement:** Common sites for psoriasis lesions to appear.
- **Scalp Involvement:** Affects the scalp in conditions like psoriasis and seborrheic dermatitis.
- **Family History:** A key factor in many dermatologic conditions, especially autoimmune diseases like psoriasis.






##### **Histopathological  Features:**

- **Melanin Incontinence:** The release of melanin from melanocytes into the dermis, seen in conditions like lichen planus.
- **Eosinophils in the Infiltrate:** Presence of eosinophils (a type of white blood cell) in the skin, often in allergic reactions or parasitic infections.
- **PNL Infiltrate (Polymorphonuclear Leukocyte Infiltrate):** Infiltration of neutrophils, commonly seen in psoriasis and some infections.
- **Fibrosis of the Papillary Dermis:** Scarring or thickening of the dermis, seen in chronic skin conditions or scars.
- **Exocytosis:** Migration of inflammatory cells into the epidermis, often seen in inflammatory skin disorders.
- **Acanthosis:** Thickening of the epidermis, often seen in psoriasis and other hyperproliferative skin diseases.
- **Hyperkeratosis:** Increased keratin production, leading to thickening of the skin, typical in psoriasis.
- **Parakeratosis:** Retention of nuclei in the stratum corneum (retention of Dead Skin cells on the skin), often seen in psoriasis and other inflammatory skin conditions.
- **Clubbing of the Rete Ridges:** A change in the shape of the dermal-epidermal junction, often seen in chronic inflammatory skin conditions.
- **Elongation of the Rete Ridges:** The extension of the epidermal ridges into the dermis, seen in chronic inflammation like psoriasis.
- **Thinning of the Suprapapillary Epidermis:** Decreased thickness of the epidermis over the dermal papillae, often seen in conditions like lichen planus.
- **Spongiform Pustule:** A collection of neutrophils in the epidermis (The Layer under the stratum Corneum), leading to pustule formation, seen in psoriasis.
- **Munro Microabscess:** Collections of neutrophils in the stratum corneum (The Top most laryer of skin which consists of Dead Cells), a hallmark of psoriasis.
- **Focal Hypergranulosis:** Thickening of the granular layer, often seen in certain inflammatory skin conditions.
- **Disappearance of the Granular Layer:** Loss of the granular layer (Deeper layer consist of the Keratin) of the epidermis, typical in acute inflammatory disorders.
- **Vacuolisation and Damage of Basal Layer:** Destruction of basal keratinocytes, typically seen in interface dermatitis (e.g., lupus erythematosus).
- **Spongiosis:** Intercellular edema (swelling in tiny spaces of your cells with flueids) in the epidermis, seen in conditions like eczema.
- **Saw-tooth Appearance of Rete:** Characteristic histological feature of lichen planus, with sharply pointed rete ridges.
- **Follicular Horn Plug:** A keratin plug in hair follicles, seen in conditions like keratosis pilaris or psoriasis.
- **Perifollicular Parakeratosis:** Parakeratosis around hair follicles, often seen in psoriasis.
- **Inflammatory Mononuclear Infiltrate:** Presence of mononuclear cells (like lymphocytes) in the dermis, seen in many inflammatory skin diseases.
- **Band-like Infiltrate:** A linear band of inflammatory cells at the dermal-epidermal junction, typical of conditions like lupus erythematosus.

### EDA
**Exploratory Data Analysis (EDA)** is a statistical method for analyzing data sets to identify patterns, relationships, and anomalies. It's often the first step in data analysis, and can help determine how to best manipulate data to get answers.
- The EDA metods used here are
  1. **Countplot :** A countplot is a type of bar plot in Seaborn used to show the counts of observations in each categorical bin using bars. It is a convenient way to visualize the distribution of a categorical variable.
  2. **Histogram :** A histogram is a type of plot used to display the distribution of a numeric variable. It divides the range of the variable into bins (intervals) and counts the number of observations in each bin. This makes it easy to see patterns such as skewness, modality, and the presence of outliers.

##### Univariate Analysis :

```python
n_subplot= len(Clinical_features.columns)
n_cols = 3  # Number of subplots per row
n_rows = math.ceil(n_subplot / n_cols)  # Calculate required rows

# Create subplots
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(18, n_rows * 4))
axes = axes.flatten()  # Flatten the axes for easy iteration

# Iterate through each column
for idx, i in enumerate(Clinical_features.columns):
    # Plotting bar graph in the appropriate subplot
    ax = sns.countplot(data=Clinical_features, x=i, palette='muted', ax=axes[idx])
    total = len(Clinical_features)
    
    # Add percentages on top of each bar
    for p in ax.patches:
        count = p.get_height()  # Bar height (count)
        percentage = f'{(count / total * 100):.1f}%'  # Calculate percentage
        x = p.get_x() + p.get_width() / 2  # Bar's center
        y = p.get_height()  # Bar's height
        ax.annotate(percentage, (x, y), ha='center', va='bottom')
    
    # Add title and labels
    ax.set_title(f'Countplot for {i}')
    ax.set_xlabel(i)
    ax.set_ylabel('Count')

# Remove empty subplots if there are any
for j in range(idx + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Feature Selection using Chi_squared Test 

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

# SelectKBest with Chi-Square
chi2_selector = SelectKBest(chi2, k='all')  # You can choose 'k' to select the top 'k' features
X_selected = chi2_selector.fit_transform(Data, Data["class"])

# Get the p-values for each feature
p_values = chi2_selector.pvalues_

# Get the feature names
feature_names = Data.columns

# Create a DataFrame to view the feature scores and p-values
feature_scores = pd.DataFrame({
    'Feature': feature_names,
    'Chi-Square Score': chi2_selector.scores_,
    'p-value': p_values
})

# Display the result
print(feature_scores)
```

**As Every features are purely categorical with medical data so,every features gives unique information which make us to consider every column**

```python
# splitting Predictors and target variable
x=Data.drop('class',axis=1)
y=Data['class']
```

```python
# unique values in target valuable
y.value_counts()
```

```python
# Spliting Data into training and testing sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

### Balancing Data Using SMOTE :

```python
# Balancing done by Oversampling method called SMOTE
smote= SMOTE(random_state=42)
x_smote,y_smote=smote.fit_resample(x,y)
print(y_smote.value_counts())
xsm_train,xsm_test,ysm_train,ysm_test=train_test_split(x_smote,y_smote,test_size=0.2,random_state=42)
```
