### Domain Analysis

- Erythemato-squamous diseases (ESDs) are a group of common skin conditions that cause inflammation and scaling, usually on the face and scalp.

- The six important types of ESD are including **psoriasis, seborrheic dermatitis, lichen planus, pityriasis rosea, chronic dermatitis and pityriasis rubra pilaris**. 

- For instance, the psoriasis is the most common disease of ESD which approximately, **2 to 3% of population** are affected.

- The skin appears red due to the shedding of skin cells and other symptoms include itching, scaling, and involvement of the oral mucosa.

- ESDs are often caused by genetic or environmental factors.

- All of them have similar clinical and histopathology features such as erythema and scaling which makes it difficult to detect them even with biopsy, biopsy is necessary to achieve a correct diagnosis. 

- Usually, clinicians detects these diseases using some clinical features, such as the degree of erythema and scaling, border of lesion, the presence of itching, koebner phenomenon, the formation of papule, involvement of oral mucosa, elbows, knees and the scalp and family history

- The six phenotypes of Erythmato-squamous Diseases are the major challenge in dermatology.

- These diseases are frequently seen in the outpatient department of dermatology. 

- All of ESD disease are same in terms of erythema and scaling, but when examine carefully some patient have the typical clinical feature of disease.

- Due to overlapping between sign and symptom of ESD, differential diagnosis is very difficult. 

- Automatic detection of these group of disease could help physician for decision making.

***Description of the features available in the dataset***

**Clinical Features :**

1. **Erythema :** Redness of the skin due to increased blood flow, often seen in inflammatory conditions.

2. **Scaling :** The shedding of dead skin cells, creating flakes on the surface, common in psoriasis, seborrheic dermatitis, etc

3. **Definite borders :** Clear, well-defined boundaries of lesions, often seen in conditions like psoriasis or ringworm.

4. **Itching :** A common symptom of many skin conditions, including eczema and psoriasis.

5. **Koebner Phenomenon :** The appearance of new skin lesions at sites of trauma or injury in patients with certain skin disorders like psoriasis or lichen planus.

6. **Polygonal Papules :** Small, raised, flat-topped lesions with a polygonal shape, seen in conditions like lichen planus.

7. **Follicular Papules :** Raised bumps that appear around hair follicles, commonly seen in acne or keratosis pilaris.

8. **Oral Mucosal Involvement :** Lesions that appear on the mucosal surfaces inside the mouth, such as in lichen planus.

9. **Knee and Elbow Involvement :** Common sites for psoriasis lesions to appear.

10. **Scalp Involvement :** Affects the scalp in conditions like psoriasis and seborrheic dermatitis.

11. **Family History :** A key factor in many dermatologic conditions, especially autoimmune diseases like psoriasis.





**Histopathological  Features :**

1. **Melanin Incontinence :** The release of melanin from melanocytes into the dermis, seen in conditions like lichen planus.

2. **Eosinophils in the Infiltrate :** Presence of eosinophils (a type of white blood cell) in the skin, often in allergic reactions or parasitic infections.

3. **PNL Infiltrate (Polymorphonuclear Leukocyte Infiltrate) :** Infiltration of neutrophils, commonly seen in psoriasis and some infections.

4. **Fibrosis of the Papillary Dermis :** Scarring or thickening of the dermis, seen in chronic skin conditions or scars.

5. **Exocytosis :** Migration of inflammatory cells into the epidermis, often seen in inflammatory skin disorders.

6. **Acanthosis :** Thickening of the epidermis, often seen in psoriasis and other hyperproliferative skin diseases.

7. **Hyperkeratosis :** Increased keratin production, leading to thickening of the skin, typical in psoriasis.

8. **Parakeratosis :** Retention of nuclei in the stratum corneum (retention of Dead Skin cells on the skin), often seen in psoriasis and other inflammatory skin conditions.

9. **Clubbing of the Rete Ridges :** A change in the shape of the dermal-epidermal junction, often seen in chronic inflammatory skin conditions.

10. **Elongation of the Rete Ridges :**  Rete ridges interlock with dermal papillae (upward projections of the dermis) to form a strong connection between the epidermis and dermis, preventing the skin from shearing or separating under stress. The extension of the epidermal ridges into the dermis, seen in chronic inflammation like psoriasis.

11. **Thinning of the Suprapapillary Epidermis :** Decreased thickness of the epidermis over the dermal papillae, often seen in conditions like lichen planus.

12. **Spongiform Pustule :** A collection of neutrophils in the epidermis (The Layer under the Stratum corneum), leading to pustule formation, seen in psoriasis.

13. **Munro Microabscess :** Collections of neutrophils in the stratum corneum (The Top most laryer of skin which consists of Dead Cells), a hallmark of psoriasis.

14. **Focal Hypergranulosis :** Thickening of the granular layer, often seen in certain inflammatory skin conditions.

15. **Disappearance of the Granular Layer :** Loss of the granular layer (Deeper layer consist of the Keratin) of the epidermis, typical in acute inflammatory disorders.

16. **Vacuolisation and Damage of Basal Layer :** Destruction of basal keratinocytes, typically seen in interface dermatitis (e.g., lupus erythematosus).

17. **Spongiosis :** Intercellular edema (swelling in tiny spaces of your cells with flueids) in the epidermis, seen in conditions like eczema.

18. **Saw-tooth Appearance of Rete :** Characteristic histological feature of lichen planus, with sharply pointed rete ridges.

19. **Follicular Horn Plug :** A keratin plug in hair follicles, seen in conditions like keratosis pilaris or psoriasis.

20. **Perifollicular Parakeratosis :** Parakeratosis around hair follicles, often seen in psoriasis.

21. **Inflammatory Mononuclear Infiltrate :** Presence of mononuclear cells (like lymphocytes) in the dermis, seen in many inflammatory skin diseases.

22. **Band-like Infiltrate :** A linear band of inflammatory cells at the dermal-epidermal junction, typical of conditions like lupus erythematosus.

## Exploratory Data Anlysis(EDA)

**Exploratory Data Analysis (EDA)** is a statistical method for analyzing data sets to identify patterns, relationships, and anomalies. It's often the first step in data analysis, and can help determine how to best manipulate data to get answers.
- The EDA metods used here are
  1. **Countplot :** A countplot is a type of bar plot in Seaborn used to show the counts of observations in each categorical bin using bars. It is a convenient way to visualize the distribution of a categorical variable.
  2. **Histogram :** A histogram is a type of plot used to display the distribution of a numeric variable. It divides the range of the variable into bins (intervals) and counts the number of observations in each bin. This makes it easy to see patterns such as skewness, modality, and the presence of outliers.

```python
## Univariate analysis for clinical_features_data

# Number of columns to analyze
num_columns = len(clinical_features_data.columns)

# Create subplots
fig, axes = plt.subplots(nrows=(num_columns + 1) // 2, ncols=2, figsize=(16, num_columns * 3))
axes = axes.flatten()  # Flatten the axes for easy iteration

# Iterate through each column
for idx, i in enumerate(clinical_features_data.columns):
    
    # Plotting bar graph in the appropriate subplot
    ax = sns.countplot(data=clinical_features_data, x=i, palette='muted', ax=axes[idx])
    total = len(clinical_features_data)
    for p in ax.patches:
        count = p.get_height()  # Bar height (count)
        percentage = f'{(count / total * 100):.1f}%'  # Calculate percentage
        x = p.get_x() + p.get_width() / 2  # Bar's center
        y = p.get_height()  # Bar's hein ght
        ax.annotate(percentage, (x, y), ha='center', va='bottom')
    ax.set_title(f'Countplot for {i}')
    ax.set_xlabel(i)
    ax.set_ylabel('Count')

# Remove empty subplots if any
for j in range(idx + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```
